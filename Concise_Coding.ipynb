# Summons Data

```r
# Getting summons data framne
summons <- read_csv("NYPD_Summons_Data.csv")

subway_summons_check <- function(column){
    n <- length(column)
    res <- rep(FALSE,n)
    for (i in 1:n)
        if (column[i] == "1050.4(A)"){
            res[i] <- TRUE
        }
   return(res)
} 
    
summons$IS_SUBWAY_SUMMONS <- subway_summons_check(summons$LAW_SECTION_NUMBER)
head(summons)
```

```r
NA_homicide <- function(column) {
    n <- length(column)
    res <- column
    for(i in 1:n){
        if (is.na(column[i]) == TRUE){
            res[i] <- "UNKNOWN"
}
}
    return(res)
}

summons$RACE <- NA_homicide(summons$RACE)
```

```r
table(summons$BORO) #CHANGING NA'S TO UNKNOWN INCREASED UNKNOWN BY 25,000
```

```r
hispanic_assimilation <- function(column){
    
    n <- length(column)
    res <- rep(0,n)
    for (i in 1:n) {
        if (column[i] == "WHITE HISPANIC"){
            res[i] <- "HISPANIC"
}               
            else if (column[i] == "BLACK HISPANIC"){
                res[i] <- "HISPANIC"
}
                else if (column[i] == "OTHER"){
                    res[i] <- "UNKNOWN"
}
                    else{
                        res[i] <- column[i]
}
}
        
   return(res)

}

summons$RACE <- hispanic_assimilation(summons$RACE)
```

# Arrest Data

```r
#Getting 
arrests <- read_csv("NYPD_Arrest_Data.csv")

arrest_check <- function(column){
    n <- length(column)
    res <- rep(FALSE,n)
    for (i in 1:n)
        if (column[i] == "PL 1651503"){
            res[i] <- TRUE
        }
   return(res)
} 

arrests$IS_FARE_ARREST <- arrest_check(arrests$LAW_CODE)
```

```r
hispanic_assimilation <- function(column){
    
    n <- length(column)
    res <- rep(0,n)
    for (i in 1:n) {
        if (column[i] == "WHITE HISPANIC"){
            res[i] <- "HISPANIC"
}               
            else if (column[i] == "BLACK HISPANIC"){
                res[i] <- "HISPANIC"
}
                else{
                    res[i] <- column[i]
}
}
        
   return(res)

}

arrests$PERP_RACE <- hispanic_assimilation(arrests$PERP_RACE)
```

```r
fare_arrests <- filter(arrests, IS_FARE_ARREST == TRUE)
arrests_by_race <- group_by(fare_arrests, PERP_RACE)
sum_race_arrests <- summarise(arrests_by_race, count=n())

#This is a crucial data frame moving forward
NYPD_race_arrests <- arrange(sum_race_arrests, PERP_RACE)
print(NYPD_race_arrests)
```
